#### Plotting Boxplot for all the independent variables for checking outliers

```python
for i in col:
    ax ,plot1 = plt.subplots(figsize = (4,4))

    plot1.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
    
```

```python
# As can be seen there are outliers in many variables, so now will treat then using IQR method
```

```python
for i in col:
    Q1 = np.percentile(df[i], 25,interpolation = 'midpoint') # Quartile 1 value
    Q3 = np.percentile(df[i], 75,interpolation = 'midpoint') # Quartile 3 value
    IQR = Q3 - Q1 # Getting the interquartile range value
    #upper fence
    uf = Q3+1.5*IQR
    
    # lower fence
    lf = Q1-1.5*IQR
    
    #Taking the indices of the values which are not present in between upper and lower fence
    upper = np.where(df[i] >= uf) # for upper fence
    lower = np.where(df[i] <= lf) # for lower fence
 
    #Removing the Outliers
    df.drop(upper[0], inplace = True)
    df.drop(lower[0], inplace = True)
    df.reset_index(inplace = True,drop=True)
```

### Encoding of categorical variable i.e. Diagnosis

```python
# Checking for the Unique Elements of Diagnosis column
df.diagnosis.unique
```

```python
# As it has only 2 unique values i.e. M and B, we can do label encoding manually by making dictionary and map()
```

```python
tempdict = {'M':1,'B':0}
df['diagnosis'] = df['diagnosis'].map(tempdict)
df['diagnosis']
```

#### Taking the Target values into y variables

```python
X = df.drop(['diagnosis'],axis = 1)
y = df['diagnosis']
```

### Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

#### Splitting data into Train and Test

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

#### Model and Fit

```python
GNB = GaussianNB()
GNB.fit(X_train,y_train)
y_pred = GNB.predict(X_test)
```

##### Score

```python
print(accuracy_score(y_test,y_pred))
```
